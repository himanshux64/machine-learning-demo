```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import seaborn as sns
```

```python
df=pd.read_csv('dataset/titanic.csv')
df.head()
```

```python

data=df.drop(columns=['PassengerId','Name','Sex','Ticket','Fare','Cabin','Embarked'],axis=True)
data.head()
```

```python
data.shape
```

```python
data=data.dropna()
data
```

```python
data.isnull().sum()
```

```python
X=data.iloc[:,1:]
y=data.iloc[:,0]
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

# Applying Feature Constructing

```python
X['Family_size']=X['SibSp']+X['Parch']+1
```

```python
X.head()
```

```python
def myFucn(num):
    if num ==1:
        # alone
        return 0
    elif num >1 and num <=4:
        #small family 
        return 1
    else:
        # large Family
        return 2
```

```python
print(myFucn(5))
```

```python
X['Family_type']=X['Family_size'].apply(myFunc)
```
