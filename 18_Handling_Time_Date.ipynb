# 1. Working date

```python
date['date']=pd.to_datetime(date['date'])
date.info()
```

```python
# Extracing year
date['date_year']=date['date'].dt.year  # dt means datetime
date.sample(6)
```

```python
# Exreacting month
date['date_month']=date['date'].dt.month
date.sample(6)
```

```python
# Extracting the name Month
date['date_month_name']=date['date'].dt.month_name()
date.sample(6)
```

```python
# Extracing day
date['date_day']=date['date'].dt.day
date.sample(6)
```

```python
# Extracing day name
date['date_day_name']=date['date'].dt.day_name()
date.sample(6)
```
