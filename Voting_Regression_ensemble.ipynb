```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
```

```python
dg=load_digits()
```

```python
df=pd.DataFrame(dg)
```

```python
sns.pairplot(dg,hue='target')
```

```python
X=df.loc[:,['area','town_robinsville','town_west windsor']]
y=df.loc[:,['price']]
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
cl1=LinearRegression()
cl2=KNeighborsRegressor()
cl3=RandomForestRegressor()
```

```python
estimators=[("lr",cl1),("knn",cl2),("rf",cl3)]
```

```python
for estimator in estimators:
    x=cross_val_score(estimator[1],X,y,cv=10)
    print(estimator[0],np.round(np.mean(x),2))
```
