```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,root_mean_squared_error
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import Lasso
```

```python
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=15)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
plt.scatter(X,y)
from sklearn.linear_model import LinearRegression
re=LinearRegression()
re.fit(X_train,y_train)
```

```python
y_re=re.predict(X_test)
```

```python
print(r2_score(y_re,y_test))
print(root_mean_squared_error(y_re,y_test))
```

```python
l=Lasso()
```

```python
l.fit(X_train,y_train)
y_pre=l.predict(X_test)
```

```python
print(r2_score(y_pre,y_test))
print(root_mean_squared_error(y_pre,y_test))
```

 # Show increase Alpha hyper parm Lasso

```python
alph=[0,1,5,10,15,30]
plt.figure(figsize=(12,6))
plt.scatter(X,y)
for i in alph:
    L=Lasso(alpha=i)
    L.fit(X_train,y_train)
    plt.plot(X_test,L.predict(X_test),label='alpha={}'.format(i))
plt.legend()
plt.show()
```
