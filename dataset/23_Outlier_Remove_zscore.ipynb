```python
data=df.drop(columns=['Unnamed: 0'],axis=1)
data.sample(5)
```

```python
sns.displot(data['iq'],kind='kde')
```

```python
print("mean of cgpa:",data['cgpa'].mean())
print("Std of cgpa :",data['cgpa'].std())
print("max value of cgpa:",data['iq'].max())
print("min value of cgpa :",data['iq'].min())
```

```python
 # finding boundary value 
print("Highest allowed :",data['cgpa'].mean()+3*data['cgpa'].std())
print("Lowest allowed :",data['cgpa'].mean()-3*data['cgpa'].std())
```
