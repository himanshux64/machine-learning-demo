```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer,KNNImputer
```

```python
df=pd.read_csv('dataset/titanic.csv',usecols=['Age','Pclass','Fare','Survived'])
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
y_train
```

# KNN Imputer

```python
knn=KNNImputer()
X_train_knn=knn.fit_transform(X_train)
X_test_knn=knn.transform(X_test)
```

```python
X_test_knn
```

```python
pd.DataFrame(X_test_knn,columns=X_train.columns)
```

```python
lr=LogisticRegression()
lr.fit(X_train_knn,y_train)
```

```python
y_pred=lr.predict(X_test_knn)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

 # SImpleImputer 

```python
si=SimpleImputer()
X_train_si=si.fit_transform(X_train)
X_test_si=si.transform(X_test)
```

```python
X_train_si
```

```python
lr1=LogisticRegression()
lr1.fit(X_train_si,y_train)
```

```python
y_pred1=lr1.predict(X_test_si)
accuracy_score(y_test,y_pred1)
```
