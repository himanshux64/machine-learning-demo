```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
dia=load_diabetes()
```

```python
df=pd.DataFrame(dia.data,columns=dia.feature_names)
```

```python
df.head()
```

```python
df['target']=dia.target
```

```python
df.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:10],df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
lr=LinearRegression()
kn=KNeighborsRegressor()
dr=DecisionTreeRegressor()
```

```python
lr.fit(X_train,y_train)
kn.fit(X_train,y_train)
dr.fit(X_train,y_train)
```

```python
y_pred1=lr.predict(X_test)
y_pred2=kn.predict(X_test)
y_pred3=dr.predict(X_test)
```

```python
print("R2 Score of Lr",r2_score(y_test,y_pred1))
print("R2 Score of KNN",r2_score(y_test,y_pred2))
print("R2 Score of DR",r2_score(y_test,y_pred3))
```

```python
from sklearn.ensemble import BaggingRegressor
```

```python
bg=BaggingRegressor(random_state=1)
bg.fit(X_train,y_train)
```

```python
y_bg=bg.predict(X_test)
```

```python
r2_score(y_test,y_bg)
```

```python
print("Training Coefficient of r2 : %.3f"%bg.score(X_train,y_train))
print("Test Coefficient of r2 : %.3f"%bg.score(X_test,y_test))
```
