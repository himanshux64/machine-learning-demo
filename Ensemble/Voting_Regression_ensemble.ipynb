```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
```

```python
digit=load_digits()
```

```python
df=pd.DataFrame(digit.data,columns=digit.feature_names)
df.head()
```

```python
df['target']=digit.target
```

```python
df.head()
```

```python
df.info()
```

```python
X=df.iloc[:,0:64]
y=df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
cl1=LinearRegression()
cl2=KNeighborsRegressor()
cl3=RandomForestRegressor()
```

```python
estimators=[("lr",cl1),("knn",cl2),("rf",cl3)]
```

```python
for estimator in estimators:
    x=cross_val_score(estimator[1],X,y,cv=10,scoring='r2')
    print(estimator[0],np.round(np.mean(x),2))
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vc=VotingRegressor(estimators)
score=cross_val_score(vc,X,y,scoring='r2',cv=10)
print("voting regressor",np.round(np.mean(score),2))
```

# Same Algorithm

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dr1=DecisionTreeRegressor(max_depth=1)
dr2=DecisionTreeRegressor(max_depth=2)
dr3=DecisionTreeRegressor(max_depth=3)
dr4=DecisionTreeRegressor(max_depth=4)
dr5=DecisionTreeRegressor(max_depth=5)
```

```python
est=[('dr1',dr1),('dr2',dr2),('dr3',dr3),('dr4',dr4),('dr5',dr5)]
```

```python
for es in est:
    s=cross_val_score(es[1],X,y,scoring='r2',cv=10)
    print(es[0],np.round(np.mean(s),2))
```

```python
vr=VotingRegressor(est)
sc=cross_val_score(vr,X,y,scoring='r2',cv=10)
print("voting regressor",np.round(np.mean(sc),2))
```
