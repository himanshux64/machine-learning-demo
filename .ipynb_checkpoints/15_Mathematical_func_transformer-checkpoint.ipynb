```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import FunctionTransformer
```

```python
df=pd.read_csv('dataset/titanic.csv',usecols=['Age','Fare','Survived'])
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Age']
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=42)
```

```python
X_train
```

# 1. Pdf and QQ-Plot

# For Age  >>>>>>>

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.distplot(X_train['Age'])
plt.title("Age PDF")

plt.subplot(122)
stats.probplot(X_train['Age'],dist='norm',plot=plt)
plt.title("Age QQ-Plot")

plt.show()
          
```

# For Fare>>>>>

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.distplot(X_train['Fare'])
plt.title("Fare PDF")

plt.subplot(122)
stats.probplot(X_train['Fare'],dist='norm',plot=plt)
plt.title("Fare QQ-plt")

plt.show()
```

```python
lr=LogisticRegression()
dtc=DecisionTreeClassifier()
```

```python
from sklearn.metrics import accuracy_score
```

```python
lr.fit(X_train,y_train)
dtc.fit(X_train,y_train)

y_pred_lr=lr.predict(X_test)
y_pred_dtc=dtc.predict(X_test)

print("Acurracy of Lr :",accuracy_score(y_pred_lr,y_test))
print("Acurracy of dtc :",accuracy_score(y_pred_dtc,y_test))
```

# Using Function Transformer

```python
trf=FunctionTransformer(func=np.log1p)
X_train_trans=trf.fit_transform(X_train)
X_test_trans=trf.transform(X_test)
```

```python
lr1=LogisticRegression()
dtc2=DecisionTreeClassifier()
```

```python
lr1.fit(X_train_trans,y_train)
dtc2.fit(X_train_trans,y_train)

y_pred_lr1=lr.predict(X_test_trans)
y_pred_dtc2=dtc.predict(X_test_trans)

print("Acurracy of Lr :",accuracy_score(y_pred_lr1,y_test))
print("Acurracy of dtc :",accuracy_score(y_pred_dtc2,y_test))
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
stats.probplot(X_train['Fare'],dist='norm',plot=plt)
plt.title("fare before Log")

plt.subplot(122)
stats.probplot(X_train_trans['Fare'],dist='norm',plot=plt)
plt.title("fare After Log")

plt.show()

```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
stats.probplot(X_train['Age'],dist='norm',plot=plt)
plt.title("Age before Log")

plt.subplot(122)
stats.probplot(X_train_trans['Age'],dist='norm',plot=plt)
plt.title("Age After Log")

plt.show()

```
